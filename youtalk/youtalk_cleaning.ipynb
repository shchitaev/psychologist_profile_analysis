```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data/youtalk_list_anon.csv", index_col=0)
```

```python
df["features"] = df["features"].apply(lambda x: x.split("\n"))
```

```python
df["features"]
```

```python
import ast
```

```python
df["educations"] = df["educations"].apply(lambda x: ast.literal_eval(x))
```

```python
df["educations"][0]
```

```python
def edu_count(x):
    ed_c = 0
    for i in x:
        if i["type"] == "education":
            ed_c += 1
    return ed_c
```

```python
df["education_counts"] = df["educations"].apply(edu_count)
```

```python
def training_count(x):
    ed_c = 0
    for i in x:
        if i["type"] == "training":
            ed_c += 1
    return ed_c
```

```python
df["training_counts"] = df["educations"].apply(training_count)
```

```python
df["training_counts"]
```

```python
list(df["approaches"].sample(2))
```

```python
df["approaches"] = df["approaches"].apply(lambda x: ast.literal_eval(x))
```

```python
def get_approaches(x):
    approaches_list = []
    for i in x:
        approaches_list.append(i["name"])
    return approaches_list
```

```python
df["approaches"] = df["approaches"].apply(get_approaches)
```

```python
df["professions"] = df["professions"].apply(lambda x: ast.literal_eval(x))
```

```python
df["professions"]
```

```python
def professions_name(x):
    professions_list = []
    for i in x:
        professions_list.append(i["profession_name"])
    return professions_list
```

```python
df["professions"] = df["professions"].apply(professions_name)
```

```python
df["professions"].value_counts()
```

```python
df["professions"] = df["professions"].apply(lambda x: x if len(x) >= 1 else np.nan)
```

```python
df.T
```

```python
df["work_areas"] = df["work_areas"].apply(lambda x: ast.literal_eval(x))
```

```python
df["non_working_areas"] = df["non_working_areas"].apply(lambda x: ast.literal_eval(x))
```

```python
df.drop(["profession"], axis=1, inplace=True)
```

```python
df.T
```

```python
def areas_name(x):
    areas_list = []
    for i in x:
        areas_list.append(i["tag"])
    return areas_list
```

```python
df["work_areas"] = df["work_areas"].apply(areas_name)
```

```python
df["non_working_areas"] = df["non_working_areas"].apply(areas_name)
```
